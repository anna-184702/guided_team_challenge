```python
# Import libraries
from helper_functions import *
import numpy as np 
from matplotlib import pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

import gpflow
from gpflow.utilities import print_summary
```

```python
# Read in data
data_heating, data_cooling, target_heating_elec, target_cooling_elec = read_in_hc_data()
# Scale the features
sc = StandardScaler()
data_heating = sc.fit_transform(data_heating)
```

```python
k = gpflow.kernels.Matern52()
```

```python
m = gpflow.models.GPR(data=(data_heating, target_heating_elec), kernel=k, mean_function=None)
```

```python
m.likelihood.variance.assign(0.01)
m.kernel.lengthscale.assign(0.3)
```

```python
opt = gpflow.optimizers.Scipy()
```

```python
def objective_closure():
    return - m.log_marginal_likelihood()

opt_logs = opt.minimize(objective_closure,
                        m.trainable_variables,
                        options=dict(maxiter=100))
print_summary(m)
```
